```python
import pandas as pd

df = pd.read_csv("/content/data.csv")

df = df[["city", "area"]].dropna()
df["city"] = df["city"].astype(str).str.strip()
df["area"] = df["area"].astype(str).str.strip()

df = df.drop_duplicates()
network = df.rename(columns={
    "city": "Source",
    "area": "Target"
})
network.to_csv("/content/gephi_city_area_network.csv", index=False)

network
```
